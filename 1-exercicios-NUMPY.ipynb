### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np
```

```
arr1 = np.random.randint(1,51,10) #de 1 a 51 para pegar o 50 (tipo o range)
media1 = np.mean(arr1)
dp1 = np.std(arr1)
print("A média é {} e o desvio padrão é {}".format(media1,dp1))
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arr2 = np.random.rand(10)
arr2
```

```
arr2[arr2 <= 0.5] = np.sqrt(arr2[arr2 <= 0.5])

print(arr2)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz3 = np.random.rand(3, 3, 3)
print(matriz3)
max3 = np.amax(matriz3, axis = 1) #trabalhar com amax quando quero obter um max de um 2D de um 3D
print("Valores maximos de cada linha de cada matriz 2D: ", max3)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz4 = np.random.rand(5, 5)
matriz4
```

```
media4 = np.mean(matriz4, axis = 1) #axis = 1 é a linha, axis = 0 é coluna
print("A média de cada linha da matriz4 é: ",media4)

matriz_menos_media = matriz4 - media4[:, np.newaxis] #np.newaxis transformando a linha da media em coluna para subtrair da matriz original
print("Matriz menos a média: ", matriz_menos_media)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz5 = np.random.rand(4, 4)
matriz5
```

```
diag_principal = np.diag(matriz5)
diag_secund = np.diag(np.fliplr(matriz5)) #função dentro da função
 
media_diagonais = np.mean([diag_principal, diag_secund])
dp_diagonais = np.std([diag_principal,diag_secund])


print("A média das diagonais é:", media_diagonais)
print("O desvio padrão das diagonais é: ", dp_diagonais)
```
